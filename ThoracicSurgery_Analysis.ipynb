```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt


targetUrl = "https://raw.githubusercontent.com/gundaminpde/2023_F_set_theory/main/ThoraricSurgery3.csv"


Data_set = pd.read_csv(targetUrl, sep=',')


X = Data_set.iloc[:,0:16]
y = Data_set.iloc[:,16]



print(Data_set.shape)
Data_set.head()
```

```python
model = Sequential()                                                  # 딥러닝 모델의 구조를 결정합니다.
model.add(Dense(30, input_dim=16, activation='relu'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])  # 딥러닝 모델을 실행합니다.
history=model.fit(X, y, epochs=70, batch_size=16)


```

```python
y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len,y_loss)
```
